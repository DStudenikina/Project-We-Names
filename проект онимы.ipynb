1 Данный проект направлен на анализ специфики ономастической лексики в романе «Мы» Евгения Замятина. Ономастическая лексика (далее – онимы) представляет собой номинации, которые именуют референты (конкретные единичные объекты). В распространненом понимании онимы – имена собсвтенные.  

В словаре Н. В. Подольской приводится подробная классификация и терминология ономастической лексики. Например, помимо широко известных классов онимов, таких как антропонимы (Маша, Вася) и топонимы (Россия, Казань), в данном словаре приводятся и другие: хрононимы (названия праздников, исторические события) и идеонимы (названия произведений искусства, названия законодательных актов и документов) [Подольская Н.В. 2 Роман «Мы» относится к жанру фантатсики по формальным признакам вымешленного мира произведения. Также он относится к жанру антиутопии по своим композиционным и идейным аспектам. Данные типы текстов предполагают высокий уровень авторского словотворчества. Ономастические единицы призваны именовать реалии вымышленного пространства, то есть номинировать несуществующие референты. Функция данной лексики – формирование образности произведения. С другой стороны, данные лексемы транслируют языковую личность автора произведения, её психолингвистические особенности и социокультурный опыт. 

Таким образом, ономастические единицы представлют интерес со следующих позиций: 
-формирование вымысла
-личность автора

```python
import re, nltk

from nltk.corpus import stopwords
stop_words = stopwords.words('russian')

from nltk.tokenize import word_tokenize
```

```python
# Токенизация корпуса по пробелам

tokens = []  # список для хранения токенов
for text in corpus:  # перебираем каждый элемент в списке корпуса
    tokens.extend(word_tokenize(text)) 

```

```python
len(tokens) 
```

```python
nltk_stopwords_ru = stopwords.words('russian') # список слов, от которх необходимо отчистить корпус
print(nltk_stopwords_ru)
```

```python

# Привести к нижнему регистру, удалить пунктацию, удалить стоп-слова
cleaned_tokens = [
    re.sub(r'[!"#$%&\'()*+,\-./:;<=>?@\[\\\]^_`{|}~…–—«»„“‘’‹›]', '', token.lower()) # пунктуация и нижний регистр
    for token in tokens
    if (token.lower() not in nltk_stopwords_ru and 
        token.strip())  # проверка на стоп-слова и непустые строки
]
cleaned_tokens = [token for token in cleaned_tokens if token] 
```

```python
# устанавливаем библиотеку для морфологического анализа
%pip install pymorphy3
```

```python
from pymorphy3 import MorphAnalyzer # инструмент для проверки частей речи
morph = MorphAnalyzer() 
```

```python
# Лемматизация корпуса


lemmatized_tokens = []
for token in cleaned_tokens:
    parsed = morph.parse(token)[0] # смотрим все разборы слова и выбираем подходящую форму
    lemma = parsed.normal_form      # начальная форма слова
    lemmatized_tokens.append(lemma)

print(lemmatized_tokens[:50])
```

```python
#добавить еще слова в список стоп-слов (в текстах предполагались иллюстрации, которые в txt приняли такой вид: 0x01 graphic 0x01. 
# Также избавились от этих символов • §, которые в частеречной разметке не маркировались как пунктуация)
nltk_stopwords_ru = nltk_stopwords_ru + ['0x01', 'graphic', '•', '§']

```

```python
#минус еще стоп-слова (из предыдущей ячейки)
list_tokens = []   
for token in (lemmatized_tokens): 
    if token not in nltk_stopwords_ru:
        list_tokens.append(token) # новый список токенов без стоп-слов.
print(list_tokens[:50])
```

```python
len(list_tokens)  # проверим, сколько токенов на данный момент
```

```python
# проверим часть речи каждого токена
for lemma in list_tokens[:10]:
    parsed = morph.parse(lemma)[0]
    print(f"{lemma}: {parsed.tag}")
```

```python
# из-за новизны и уникальности авторской ономастики многие слова (онимы) не маркирутся верно. Поэтому придется оставить все знаменательные части речи
no_parsorat_tokens = []
for lemma in list_tokens:
    parsed = morph.parse(lemma)[0]  
    if parsed.tag.POS not in {'CONJ', 'PRCL', 'INTJ', 'PNCT'}:  # убираем служебные части речи, пунктуацию
        no_parsorat_tokens.append(lemma)

print(no_parsorat_tokens[:50])
```

```python
len(no_parsorat_tokens)  # проверим, стало ли меньше токенов
```

```python
from collections import Counter # импортируем словарь для подсчета токенов
import pandas as pd # импортируем инструмент для обработки данных
```

```python
corpus1 = no_parsorat_tokens.copy()  # Создаём копию с новым именем для таблицы
```

```python
# Посчитаем абсолютную и относительную частоту всех слов в корпусе

df = pd.DataFrame(Counter(corpus1).items(), columns=['слово', 'абсолютная частота'])
df['относительная частота'] = df['абсолютная частота'] / len(corpus1)


print(df.head(10))
```

```python
df.to_csv("частоты_слов1.csv", index=False, sep=';', encoding='utf-8-sig')
```

```python

%pip install natasha

from natasha import (
    Segmenter,
    MorphVocab,

    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,

    PER,
    NamesExtractor,

    Doc
)


segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)
```

```python

with open("Замятин_Мы.txt", "r", encoding='utf-8') as file:
    my = file.read()

corpus = my.split('\n') 
all_names = []
seen_names = set() 

for i in corpus:
    if not i.strip():  
        continue
        
    doc = Doc(i)
    doc.segment(segmenter)
    doc.tag_ner(ner_tagger)
    
    persons = []
    for span in doc.spans:
        if span.type == 'PER':
            name = span.text.lower()
            if name not in seen_names:  
                persons.append(name)
                seen_names.add(name)
    
    all_names += persons 

print(all_names[:50])
```

```python
# Лемматизируем полученные имена

lemmatized_all_names = []
for token in all_names:
    parsed = morph.parse(token)[0] 
    lemma = parsed.normal_form     
    lemmatized_all_names.append(lemma)

print(lemmatized_all_names[:50])
```

```python
# Частотность ис из наташи в корпусе2 и корпусе1
df = pd.DataFrame(Counter(lemmatized_all_names).items(), columns=['слово', 'абсолютная частота'])
df['частота_по_corpus2'] = df['абсолютная частота'] / len(corpus2)
df['частота_по_corpus1'] = df['абсолютная частота'] / len(corpus1)


print(df.head(10))
```

```python
df.to_csv("частота_наташа.csv", index=False, sep=';', encoding='utf-8-sig')
```
